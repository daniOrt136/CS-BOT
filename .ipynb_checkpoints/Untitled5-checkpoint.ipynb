```python
#!/usr/bin/env python
# coding: utf-8

# In[3]:


def excel_manipulation(ruta):
    wb = xl.load_workbook(ruta)
    ws = wb.active
    
    global temp_df
    l = len(temp_df)+1

    #Changing Zoom to a sheet
    ws.sheet_view.zoomScale = 100

    #Defining the border style
    thin_border = Border(left=Side(style='thin'),
                         right=Side(style='thin'),
                         top=Side(style='thin'),
                         bottom=Side(style='thin')                    
                        )

    #applying borders to all table
    for row in ws.iter_rows():
            for cell in row:
                cell.alignment = Alignment(wrapText=False)
                cell.font = Font(name='Calibri', size=11)
                cell.border = thin_border
                cell.alignment = Alignment(vertical='center', horizontal='center')
    #modifing the header style
    for i in ws[1]:
        i.font = Font(color="00000000", name='Calibri', size=12, bold = True)

    a1 = ws['A1']
    b1 = ws['B1']
    c1 = ws['C1']
    d1 = ws['D1']
    e1 = ws['E1']
    f1 = ws['F1']
    g1 = ws['G1']
    h1 = ws['H1']
    i1 = ws['I1']
    j1 = ws['J1']
    k1 = ws['K1']
    l1 = ws['L1']
    m1 = ws['M1']
    n1 = ws['N1']
    o1 = ws['O1']
    p1 = ws['P1']
    q1 = ws['Q1']
    r1 = ws['R1']
    
    #Clean blue
    a1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    b1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    c1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    d1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    e1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    f1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    g1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    h1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    i1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    j1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    k1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    l1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    m1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")
    o1.fill = PatternFill(fgColor="00b3c4e9", fill_type = "solid")

    
    #Green
    n1.fill = PatternFill(fgColor="0079da4c", fill_type = "solid")
    p1.fill = PatternFill(fgColor="0079da4c", fill_type = "solid")
    q1.fill = PatternFill(fgColor="0079da4c", fill_type = "solid")
    r1.fill = PatternFill(fgColor="0079da4c", fill_type = "solid")
    
    #fit to the width
    col = ws.column_dimensions['A'].width = 25
    col = ws.column_dimensions['B'].width = 15
    col = ws.column_dimensions['C'].width = 9
    col = ws.column_dimensions['D'].width = 9
    col = ws.column_dimensions['E'].width = 15
    col = ws.column_dimensions['F'].width = 25
    col = ws.column_dimensions['G'].width = 35
    col = ws.column_dimensions['H'].width = 10
    col = ws.column_dimensions['I'].width = 12
    col = ws.column_dimensions['J'].width = 15
    col = ws.column_dimensions['K'].width = 18
    col = ws.column_dimensions['L'].width = 25
    col = ws.column_dimensions['M'].width = 15
    col = ws.column_dimensions['N'].width = 50
    col = ws.column_dimensions['O'].width = 55
    col = ws.column_dimensions['P'].width = 40
    col = ws.column_dimensions['Q'].width = 45
    col = ws.column_dimensions['R'].width = 30

    
    #Heigh of the header row
    ws.row_dimensions[1].height = 30
    
    #Autocomplete formul writing the formula in each cell
    starter = 0
    i = 1
    for cell in ws['O']:
        if starter == 0:
            pass
        else:
            #Adding suggestions based on the response
            exc_formula = '=IF(N{}="Line Item Delivered-Shipped (Complete Shipment)","Please provide Shipping details / Tracking # / POD in column R",IF(N{}="Original Delivery Date Remains","Please Provide comments if they are necessary in column P",IF(N{}="New Delivery Date","Please Provide new date in column Q",IF(N{}="PO Not Received","Please Provide comments if they are necessary in column P",IF(N{}="Payment Issues","Please Provide reason why in column P",IF(N{}="Line Item Needs to be Cancelled","Please Provide reason why in column P",IF(N{}="Line Item Needs to be Modified","Please Specify changes needed in column P",IF(N{}="Material Ready for Pick Up","Please send Delivery Documents Packing Slip / Commercial Invoice / Material Safety Data",""))))))))'.format(i,i,i,i,i,i,i,i)
            cell.value = '{}'.format(exc_formula) 
        starter+=1
        i+=1
    
    #Dropdown List
    dv = DataValidation(type="list", formula1='"Line Item Delivered-Shipped (Complete Shipment), Original Delivery Date Remains, New Delivery Date, PO Not Received, Payment Issues, Line Item Needs to be Cancelled, Line Item Needs to be Modified, Material Ready for Pick Up"', allow_blank=False)
    ws.add_data_validation(dv)
    
    #Adding dropdown list to the cell
    dv.add('N2:N{}'.format(l))
    
    #Protecting the sheet
    ws.protection.sheet = True
    ws.protection.password = 'CS2021'
    
    #Cells to unlock
    c_unlock = ['N', 'P', 'Q', 'R']
    
    #Unlocking needed cells
    for col in c_unlock:
        starter = 0
        for cell in ws[col]:
            if starter == 0:
                pass
            else:
                cell.protection = Protection(locked=False)
            starter+=1

    #Data validation of column Q of new date
    ds = DataValidation(type="custom",formula1='=AND(ISNUMBER(Q2),LEFT(CELL("format",Q2),1)="D")')
    ws.add_data_validation(ds)
    
    ds.add('Q2:Q{}'.format(l))

    #Format for Column P of new del date
    starter = 0
    for cell in ws['Q']:
        if starter == 0:
            pass
        else:
            cell.protection = Protection(locked=False)
            cell.number_format = "mm/dd/yyyy"
        starter+=1
    
    #Priority coloration
    starter = 0
    for cell in ws['M']:
        if starter == 0:
            pass
        else:
            if cell.value == 'Medium':
                #Yellow
                cell.fill = PatternFill(fgColor="00ffee97", fill_type = "solid")
                cell.font = Font(color="00a55127")
                
            elif cell.value == 'High':
                #Red
                cell.fill = PatternFill(fgColor="00ffc1cc", fill_type = "solid")
                cell.font = Font(color="00aa004f")
                
            elif cell.value == 'Low':
                #Red
                cell.fill = PatternFill(fgColor="00baffce", fill_type = "solid")
                cell.font = Font(color="00006800")
        starter+=1
    
    wb.save(filename=ruta)
```
